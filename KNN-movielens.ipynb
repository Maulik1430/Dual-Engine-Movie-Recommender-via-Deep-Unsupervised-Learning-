```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix
```

```python
USER_ID = 1      # Target user
K_NEIGHBORS = 15 # Number of neighbors
N_RECOMMENDATIONS = 10 # Number of recommendations
```

```python
df = pd.read_csv("/Users/mauliksindhva/Downloads/movie-lens-1m/ratings.csv")
df.sample(10)
```

```python
df_movies = pd.read_csv("/Users/mauliksindhva/Downloads/movie-lens-1m/movies.csv")
df_movies.sample(10)
```

```python
R = df.pivot(index="user_id", columns="movie_id", values="rating")
R.shape
```

```python
df_pivot.sample(5)
```

```python
R_filled = R.fillna(0)
R_sparse = csr_matrix(R_filled.values)
user_to_index = {user_id: i for i, user_id in enumerate(R.index)}
index_to_user = {i: user_id for i, user_id in enumerate(R.index)}
target_index = user_to_index[USER_ID]
```

```python
users_avg_rating = df_pivot.mean(axis=1)
len(users_avg_rating)
```

```python
df_norm = df_pivot.sub(users_avg_rating, axis='rows')
df_norm.sample(5)
```

```python
df_norm = df_norm.fillna(0)
df_norm.sample(5)
```

```python
knn_model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=K_NEIGHBORS + 1)
knn_model.fit(R_sparse)
```

```python
distances, indices = knn_model.kneighbors(R_sparse[target_index], n_neighbors=K_NEIGHBORS + 1)
```

```python
neighbor_indices = indices.flatten()[1:]
neighbor_similarities = 1 - distances.flatten()[1:]
neighbor_user_ids = [index_to_user[i] for i in neighbor_indices]
nearest_neighbors = pd.Series(neighbor_similarities, index=neighbor_user_ids)
nearest_neighbors
```

```python
user_id = 5001

df_user = df[df["user_id"] == user_id]
df_user

rated_movies = df_user["movie_id"].to_list()
len(rated_movies)
```

```python
unrated_movies = R.columns.difference(rated_movies)
```

```python
predictions = {}
neighbor_ratings = R.loc[nearest_neighbors.index, unrated_movies]

for movie_id in unrated_movies:
    ratings = neighbor_ratings[movie_id].dropna()
    
    if len(ratings) > 0:
        # Get similarities for only those neighbors who rated the movie
        valid_neighbors = ratings.index
        weights = nearest_neighbors.loc[valid_neighbors].values
        
        # Collaborative Filtering Prediction Formula: Sum(rating * weight) / Sum(|weight|)
        numerator = np.sum(ratings.values * weights)
        denominator = np.sum(np.abs(weights))

        if denominator != 0:
            predictions[movie_id] = numerator / denominator
```

```python
if predictions:
    # Convert predictions dictionary to a sorted Series
    df_predictions = pd.Series(predictions).sort_values(ascending=False)
    top_recommendations = df_predictions.head(N_RECOMMENDATIONS).index.tolist()

    # Join with movie titles and format
    recommended_movies = df_movies[df_movies['movie_id'].isin(top_recommendations)]
    final_recs = recommended_movies.set_index('movie_id').loc[top_recommendations]
    final_recs['Predicted_Rating'] = [df_predictions.loc[mid] for mid in top_recommendations]

    print(f"\n--- Top {N_RECOMMENDATIONS} Movies for User {USER_ID} (using Sklearn KNN) ---")
    print(final_recs[['title', 'Predicted_Rating']].to_markdown(index=False))
else:
    print(f"\nCould not generate recommendations for User {USER_ID}.")
```
